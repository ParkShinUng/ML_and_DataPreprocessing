```python
import pandas as pd
from sklearn.datasets import load_iris

iris_data = load_iris()

print(type(iris_data))
```

```python
print(iris_data.keys())
```

```python
# iris_data['feature_names'] : 독립변수(입력변수, 설명변수)
# 꽃받침의 길이와 폭, 꽃잎의 길이와 폭
print(iris_data['feature_names'])
print(iris_data['data'][0:20])
```

```python
# 레이블(종속변수(반응변수)) 확인
# 품종 - setosa, versicolor, virginica
print(iris_data['target_names'])
iris_data['target']
```

```python
iris_df = pd.DataFrame(iris_data['data'], columns=iris_data['feature_names']) # data 의 type : numpy.ndarray
iris_df
```

```python
# 품종을 분류하기 위해 iris의 target컬럼을 추가
iris_df['target'] = iris_data['target']
iris_df
```

```python
dict(enumerate(iris_data['target_names']))
```

* 코드 해석 : 품종을 분류하기 위해 iris_df의 target 컬럼에 있는 값(0~2)을 
* 0은 setosa, 1은 versicolor, 2는 virginica로 대체하여 iris_df['target_desc'] 컬럼에 추가한다.

```python
# 품종을 분류하기 위한 target(종속변수, 설명변수)값의 Text 추가
iris_df['target_desc'] = iris_df['target'].replace(dict(enumerate(iris_data['target_names'])))

iris_df.tail()
```
