```python
import pandas as pd

final_df = pd.read_csv('contest_data/final_data.csv')
```

```python
final_df.columns
```

```python
final_df.drop(['AAA', 'AA', 'A', 'BBB', 'BB', 'B', 'CCC', 'CC', 'C', 'D'], axis='columns', inplace=True)
```

```python
final_df
```

```python
#부도율 = 부도 / 정상 + 관심 + 관찰1 + 관찰2 + 관찰3
all_cnt = final_df['정상'] + final_df['관심'] + final_df['관찰1'] + final_df['관찰2'] + final_df['관찰3'] + final_df['부도']
all_cnt
```

```python
rate = final_df['부도'] / all_cnt
final_df['부도율'] = rate
final_df.head(10)
```

```python
0.001712 / final_df['부도율'].sum()
```

```python
all_ent_sum = final_df['정상'].sum() + final_df['관심'].sum() + final_df['관찰1'].sum() \
+ final_df['관찰2'].sum() + final_df['관찰3'].sum() + final_df['부도'].sum()

all_bank_sum = final_df['부도'].sum()
all_bank_rate = all_bank_sum / all_ent_sum

all_bank_rate_sum = final_df['부도율'].sum()

final_df['전국비례부도율'] = final_df['부도율'] / all_bank_rate_sum

final_df.head()
```

```python
final_df['전국비례부도율'].sum()
```

```python
value = final_df.loc[final_df['SIGNGU_NM'] == '평택시','부도율']
rate_val = value / final_df['부도'].sum()
rate_val
#final_df['부도율'].value_counts()
```
