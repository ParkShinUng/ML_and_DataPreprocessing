```python
import pandas as pd

excel_path = 'data/Raw 6months reading of Gas turbine.xlsx'
# sheet name 가져오기, 모든 sheet를 가져오기 위해 sheet_name을 None으로 설정
sheet_names_dict = pd.read_excel(excel_path, sheet_name=None).keys()
```

```python
# 코드 수정

import re
from datetime import date, time, datetime

# 코드 설계
# 1. sheet 읽어오기
# 2. sheet의 범위 지정해서 읽어오기
# 3. 범위 지정한 데이터 프레임 전치
# 4. Date, Shift 컬럼 제거
# 4. Year, Month, Day, Hour 컬럼 처리
# 5. datetime 컬럼 처리 (Year-Month-Day-Hour)
#  - 컬럼명 변경 df = df.rename(columns={'A': 'a'})
#  - 인덱스 행 변경 df = df.rename(index={'ONE': ,'one'})
turbin_df = pd.DataFrame()

except_list = []
for main_idx, sheet_name in enumerate(sheet_names_dict, 1):
    if main_idx % 10 == 0:
        print(f'{main_idx} 번째 실행중')

    print(f'{sheet_name} sheet reading')

    try:
        # 엑셀 파일의 시트 범위 지정해서 읽어오는 구문
        read_sheet_df = pd.read_excel(excel_path, sheet_name=sheet_name, index_col=0)
        read_sheet_df = read_sheet_df.iloc[:52, :13]

        # -----dataframe 전치
        read_sheet_df = read_sheet_df.T

        # -----Year, Month, Day 추출 후 컬럼 추가
        all_date = re.findall(r'[0-9]+', sheet_name)
        # Year
        read_sheet_df['Year'] = '20'+all_date[0][-2:]
        # Month
        read_sheet_df['Month'] = all_date[0][-4:-2]
        # Day
        read_sheet_df['Day'] = all_date[0][-6:-4]

        # -----Sheet의 Hour 추출
        hour_list = [hour.hour for hour in read_sheet_df['Signal Names / Time']]
        read_sheet_df['Hour'] = hour_list

        turbin_df = pd.concat([turbin_df, read_sheet_df], axis=0, ignore_index=True)
    except Exception as e:
        except_list.append(e)
        continue
    
# -----datetime 데이터 설정
# 1. reset_index 후 Date, Shift 컬럼 삭제
turbin_df.reset_index(inplace=True)
turbin_df.drop(columns=['index', 'Shift'], inplace=True)
# turbin_df.drop(columns='index', inplace=True)
# turbin_df.drop(columns='Shift', inplace=True)

# 2. Year / Month / Day / Hour 을 datetime 형식으로 변경하여 Signal Names / Time 컬럼에 데이터 추가
date_str_list = []
# date_str_list.append(f'{date(turbin_df["Year"][i], turbin_df["Month"][i], turbin_df["Day"][i])} {"{0.hour:2}:{0.minute:2}".format(turbin_df["Signal Names / Time"][i])}')
for i in range(len(turbin_df['Signal Names / Time'])):
    date_str = f'{turbin_df["Year"][i]}-{turbin_df["Month"][i]}-{turbin_df["Day"][i]} \
    {"{0.hour:02}:{0.minute:02}".format(time(turbin_df["Hour"][i], 0))}'
    date_str_list.append(date_str)
turbin_df['Signal Names / Time'] = date_str_list

# 3. Signal Names / Time 컬럼을 datetime 컬럼으로 이름 변경
turbin_df = turbin_df.rename(columns={'Signal Names / Time': 'datetime'})

# -----datetime 컬럼을 인덱스 컬럼으로 재지정
turbin_df.index = turbin_df['datetime']

# -----datetime 컬럼 제거
turbin_df.drop(columns='datetime', inplace=True)

print(except_list)

turbin_df.head()
```
